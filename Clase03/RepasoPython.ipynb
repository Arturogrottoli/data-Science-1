# **Métodos de Strings**

Las cadenas de texto en Python son inmutables, lo que significa que no pueden ser modificadas después de ser creadas. Sin embargo, se puede utilizar varios métodos para crear y manipular nuevas cadenas basadas en la original.

**str.upper():** Convierte todos los caracteres de la cadena a mayúsculas.

```python
string = "hola"
string.upper()
```

**str.lower():** Convierte todos los caracteres de la cadena a minúsculas.

```python
string = "HOLA"
print(string.lower())
```

**str.capitalize():** Convierte el primer carácter de la cadena a mayúscula y el resto a minúsculas.


```python
string = "hola daniela"
print(string.capitalize())
```

**str.title():** Convierte el primer carácter de cada palabra en la cadena a mayúscula.

```python
string = "hola mundo"

print(string.title())
```

**str.strip():** Elimina los espacios en blanco al principio y al final de la cadena.

```python
string = "  hola   "
```

```python
print(string.strip())
```

**str.replace(old, new):** Reemplaza todas las apariciones de old con new en la cadena.

```python
string = "hola mundo" # Reemplazar "hola" por "mundo" y "mundo por "Python"
```

```python
string1 = string.replace("hola mundo","mundo Python")
string1
```

**str.split():** Divide la cadena en una lista de palabras, usando el separator como delimitador. Si no se especifica, usa el espacio como separador.

```python
string= "hola a todos"
```

```python
var = string.split()
print(var)
```

```python
type(var)
```

**str.join(iterable):** Une los elementos de iterable en una cadena, usando la cadena actual como separador.

```python
lista = ["hola", "a", "todos"]   # (" ".join(lista)
```

```python
(" ".join(lista))
```

**str.find(substring):** Devuelve el índice de la primera aparición de substring en la cadena, o -1 si no se encuentra.

```python
string = "hola mundo"   # find("mundo")
```

```python
string.find("mundo")
```

**str.count(substring):** Cuenta cuántas veces aparece substring en la cadena.

```python
string = "hola mundo, hola a todos" # Cuantas veces aparece "hola"?
string.count(" ")
```

# **Métodos de Listas**

Las listas en Python son mutables, lo que significa que pueden ser modificadas después de ser creadas.

**list.append(elemento):** Añade elemento al final de la lista.

```python
lista = [1, 2, 3]
lista.append(True)
print(lista)
```

**list.insert(index, elemento):** Inserta elemento en la posición index.

```python
lista = [1, 2, 3]
lista.insert(0, False)
print(lista)
```

**list.remove(elemento):**Elimina la primera aparición de elemento en la lista.

```python
lista = [1, 2, 3, 2]
lista.remove(2)
print(lista)
```

**list.pop(index):** Elimina y devuelve el elemento en la posición index. Si index no se especifica, elimina el último elemento.

```python
lista = [1, 2, 3]
elemento = lista.pop()
print(elemento)
print(lista)
```

**list.clear():** Elimina todos los elementos de la lista.

```python
lista = [1, 2, 3, 4]
lista.clear()
lista
```

**list.index(elemento):** Devuelve el índice de la primera aparición de elemento en la lista.

```python
lista = [1, "hola", 3, False]
print(lista.index(1))
```

**list.count(elemento):** Cuenta cuántas veces aparece elemento en la lista.

```python
lista = [1, 2, 3, 2, 2]
print(lista.count(2))
```

**list.sort():** Ordena los elementos de la lista en orden ascendente.

```python
lista = ["manzana", "banana"]
lista.sort()
print(lista)
```

```python
lista
```

**list.sort(reverse=True):** Ordena los elementos de la lista en orden descendente.

```python
lista = [3, 1, 4, 1]
lista.sort(reverse=True)
print(lista)
```

**list.reverse():** Invierte el orden de los elementos en la lista.

```python
lista = [1, 2, 3, 2, 5]
lista.reverse()
print(lista)
```

# **Métodos de Conjuntos (Sets)**

Los conjuntos en Python son mutables, pero no permiten elementos duplicados.

**set.add(elemento)**: Añade elemento al conjunto.

```python
lista = [1, 2, 3, 4, 4, 4]
```

```python
set(lista)
```

```python
conjunto = {1, 2, 3}
conjunto.add(4)
print(conjunto)
```

**set.remove(elemento):** Elimina elemento del conjunto. Si el elemento no existe, lanza un error.

```python
conjunto = {1, 2, 3}
conjunto.remove(3)
print(conjunto)
```

**set.discard(elemento):** Elimina elemento del conjunto si existe. Si no existe, no hace nada.

```python
conjunto.discard(5)
```

```python
conjunto
```

**set.pop():** Elimina y devuelve un elemento arbitrario del conjunto.

```python
conjunto = {1, 2, 3, 4}
elemento = conjunto.pop()
#print(elemento)
print(conjunto)
```

**set.clear():** Elimina todos los elementos del conjunto.

```python
conjunto = {1, 2, 3, 4}
conjunto.clear()
conjunto
```

**set.union(other_set):** Devuelve la unión de dos conjuntos.

```python
conjunto1 = {1, 2, 3}
conjunto2 = {3, 4, 5}
union = conjunto1.union(conjunto2)
print(union)
```

**set.intersection(other_set):** Devuelve la intersección de dos conjuntos.

```python
conjunto1 = {1, 2, 3}
conjunto2 = {2, 3, 4}
interseccion = conjunto1.intersection(conjunto2)
print(interseccion)
```

**set.difference(other_set):** Devuelve la diferencia entre dos conjuntos.

```python
conjunto1 = {1, 2, 3}
conjunto2 = {2, 3, 4}
diferencia = conjunto1.difference(conjunto2)
diferencia
```

**set.symmetric_difference(other_set):** Devuelve la diferencia simétrica entre dos conjuntos (elementos que están en uno u otro, pero no en ambos).

```python
conjunto1 = {1, 2, 3}
conjunto2 = {2, 3, 4}
diferencia_simetrica = conjunto1.symmetric_difference(conjunto2)
print(diferencia_simetrica)
```

# **Métodos de Diccionarios**

Los diccionarios en Python son mutables y permiten almacenar pares clave-valor.

**dict.keys():** Devuelve una vista de todas las claves en el diccionario.

```python
diccionario = {"nombre": "Juan", "edad": 30}
print(diccionario.keys())
```

**dict.values():** Devuelve una vista de todos los valores en el diccionario.

```python
diccionario = {"nombre": "Juan", "edad": 30}
print(diccionario.values())
```

**dict.items():** Devuelve una vista de todos los pares clave-valor en el diccionario.

```python
diccionario = {"nombre": "Juan", "edad": 30}
print(diccionario.items())
```

**dict.clear():** Elimina todos los pares clave-valor del diccionario.

```python
diccionario = {"nombre": "Juan", "edad": 30}
diccionario.clear()
print(diccionario)
```

**dict.update(otro_diccionario):** Actualiza el diccionario con los pares clave-valor de otro_diccionario. Si alguna clave ya existe, su valor será reemplazado.

```python
diccionario = {"nombre": "Juan", "edad": 30}
diccionario.update({"edad": 31, "ciudad": "Buenos Aires"})
print(diccionario)
```

```python
diccionario = {"nombre": "Juan", "edad": 30}
```

```python
diccionario.update({"nombre": ("juan", "Daniela", "María")})
diccionario
```

# **Estructuras de control**

**Estructura for**

La estructura for se utiliza para iterar sobre una secuencia (como una lista, una cadena, un rango de números, etc.) y ejecutar un bloque de código para cada elemento de la secuencia.

```python
tareas = ["Desayunar", "Ir al trabajo", "Almorzar", "Hacer ejercicio", "cocinar"]

for i in tareas:
    print(tareas)
```

```python
tareas = ["Desayunar", "Ir al trabajo", "Almorzar", "Hacer ejercicio", "Dormir"]

for tarea in tareas:
    print(f"Tarea completada: {tarea}")
```

En este ejemplo, el for recorre cada elemento de la lista tareas y ejecuta el bloque de código dentro del ciclo para cada tarea.

**F string**

 Las f-strings permiten incluir expresiones dentro de una cadena utilizando llaves {}, lo que facilita la inserción de variables y expresiones directamente en el texto

```python
nombre = "Gustavo"
edad = 39
saludo = f"Mi nombre es {nombre}."
print(saludo)
```

**Uso con range():**

```python
for i in range(5):
    print(i)
```

```python
for i in range(6,8):
    print(i)
```

**Estructura while**

La estructura while ejecuta un bloque de código repetidamente mientras una condición sea verdadera. Se usa cuando no se sabe de antemano cuántas veces se debe repetir el ciclo.

```python
contador = 0

while contador <= 5:
    print(f"Contador: {contador}")
    contador += 1
```

Este ciclo while sigue ejecutándose mientras la variable contador sea menor que 5. En cada iteración, el valor de contador se incrementa en 1.

**If, Elif, Else**

La estructura if se utiliza para tomar decisiones en el programa. Evalúa una condición y, si es verdadera, ejecuta un bloque de código. Opcionalmente, se pueden usar elif y else para manejar múltiples condiciones.

```python
clima = "soleado"

if clima == "lluvioso":
    print("Lleva un paraguas.")
elif clima == "caluroso":
    print("usa protector solar")
elif clima == "soleado":
    print("No necesitas un paraguas.")
else:
    print("Revisa el clima antes de salir.")
```

# **Funciones**

Las funciones en Python son bloques de código reutilizables que realizan una tarea específica. Se definen utilizando la palabra clave **def**, seguidas del nombre de la función, paréntesis (que pueden contener parámetros), y dos puntos. El cuerpo de la función contiene las instrucciones que se ejecutarán cuando la función sea llamada.

Características de las funciones:

**Reutilización:**
Puedes definir una función una vez y llamarla múltiples veces en tu código.

**Modularidad:** Ayudan a dividir un programa en partes más pequeñas y manejables.

**Parámetros y argumentos:** Las funciones pueden tomar entradas, conocidas como parámetros, y devolver salidas, conocidas como valores de retorno.

```python
def saludar():
    print("Hola, ¿cómo estás?")
```

```python
saludar()
```

```python
def suma(a, b):







    return a + b
```

```python
suma(10,89)
```
