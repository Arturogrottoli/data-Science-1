
# ***1. Introducción a Series***

---







Una pd.Series es un arreglo unidimensional etiquetado capaz de almacenar cualquier tipo de dato (enteros, cadenas, flotantes, objetos de Python, etc.). A diferencia de una lista común, las Series están optimizadas para operaciones vectorizadas.

Componentes: Consisten en un arreglo de datos y un arreglo de etiquetas asociado, llamado índice (index).

Eje: Posee un único eje (eje 0).

**Codigo de Ejemplo:**

---

```python
import pandas as pd

# pd.Series: crea una estructura 1D con índice a partir de cualquier iterable
# index=[...]: asigna etiquetas personalizadas en vez del índice numérico 0,1,2
precios = pd.Series([10.5, 20.0, 15.75], index=['Pan', 'Leche', 'Arroz'])
print(precios)
```


# ***3. Creación de Arrays (NumPy)***

---



Los ndarray de NumPy son bloques de memoria contiguos del mismo tipo de dato (homogéneos).

Eficiencia: Al ser homogéneos, permiten que las operaciones se realicen en C (en lugar de Python), lo que las hace órdenes de magnitud más rápidas.

Broadcasting: Permiten realizar operaciones entre arreglos de diferentes formas de manera automática.


**Codigo de Ejemplo:**

---

```python
import numpy as np

# np.array(): convierte una lista Python a ndarray — misma interfaz pero almacenado en C
arr = np.array([1, 2, 3, 4, 5])

# np.arange(start, stop): genera enteros desde start hasta stop-1, igual que range()
# pero devuelve un ndarray listo para operaciones vectorizadas
arr = np.arange(1, 6)

# Operación vectorizada: * 2 multiplica TODOS los elementos sin necesidad de un bucle
# Internamente se ejecuta en C → mucho más rápido que [x*2 for x in lista]
print(arr * 2)
```

---

 NumPy provee métodos para inicializar arreglos sin necesidad de definirlos manualmente.

Atributos clave:

  .shape: Tupla con las dimensiones.

  .size: Cantidad total de elementos.

  .ndim: Número de dimensiones (rango).

  .dtype: Tipo de dato interno (ej. float64, int32).

# ***5. Álgebra Lineal con NumPy***

---

NumPy implementa bibliotecas de bajo nivel (como BLAS y LAPACK) para cálculos matriciales.

-Producto Punto (np.dot): Realiza la suma de productos de los elementos correspondientes. En 2D, equivale a la multiplicación de matrices tradicional.

-Importancia: Es la base de los algoritmos de Machine Learning, como la regresión lineal y las redes neuronales.

**Codigo de Ejemplo:**

---


```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# np.dot(a, b): producto matricial → resultado[i,j] = suma(fila_i(a) × col_j(b))
# Ejemplo: resultado[0,0] = 1*5 + 2*7 = 19
#          resultado[0,1] = 1*6 + 2*8 = 22
# Equivalente moderno y más legible: a @ b (operador @ desde Python 3.5)
print(np.dot(a, b))
```



---


# ***8. Manejo de Valores Faltantes (NaN)***

---

En Pandas, los valores faltantes se representan como NaN (Not a Number), que técnicamente es un flotante.

Estrategias:

-Imputación: Rellenar con fillna() (usando la media, mediana o un valor constante).

-Eliminación: Usar dropna() cuando la cantidad de datos faltantes es insignificante o el sesgo de eliminarlos es bajo.

**Codigo de Ejemplo:**

---



```python
import numpy as np

# np.nan: valor especial float que Pandas usa para representar datos ausentes
# Al crearlo con np.nan, Pandas lo reconoce automáticamente como "faltante"
df_nulo = pd.DataFrame({'Ventas': [100, np.nan, 200]})

# fillna(valor): devuelve una COPIA con todos los NaN reemplazados por el valor dado
# Aquí reemplazamos NaN por 0; también se puede usar la media: fillna(df_nulo['Ventas'].mean())
print(df_nulo.fillna(0))
```



---


Práctica: Crea un DataFrame con algunos valores np.nan y elimínalos usando .dropna().


---



```python
# PRÁCTICA: DataFrame con NaN dispersos en distintas filas y columnas
df_nans = pd.DataFrame({'A': [1, np.nan, 3], 'B': [4, 5, np.nan]})

# dropna(): elimina filas que tengan AL MENOS un NaN (how='any' por defecto)
# Resultado: solo queda la fila 0 (A=1, B=4) — las filas 1 y 2 se descartan
# Parámetros útiles: how='all' (solo si TODOS son NaN), subset=['A'] (solo revisar col A)
print(df_nans.dropna())
```


# ***9. Agrupamiento de Datos (Group By)***

---

Sigue el proceso Split-Apply-Combine:

Split: Divide los datos en grupos basados en criterios.

Apply: Aplica una función (suma, promedio, conteo) a cada grupo de forma independiente.

Combine: Combina los resultados en una nueva estructura.

**Codigo de Ejemplo:**

---






```python
# DataFrame de ventas con dos tiendas y cuatro registros
ventas = pd.DataFrame({'Tienda': ['A', 'B', 'A', 'B'], 'Monto': [10, 20, 15, 30]})

# groupby('Tienda'): SPLIT — divide el df en grupos por valor único de 'Tienda'
# .sum(): APPLY — suma los montos dentro de cada grupo
# COMBINE — devuelve un nuevo df con un renglón por grupo
# Tienda A: 10 + 15 = 25  |  Tienda B: 20 + 30 = 50
print(ventas.groupby('Tienda').sum())
```



---


Práctica: Con el DataFrame de ventas, calcula el promedio (.mean()) de ventas por tienda.


---



```python
# PRÁCTICA: .mean() calcula el promedio de 'Monto' por grupo
# Tienda A: (10 + 15) / 2 = 12.5  |  Tienda B: (20 + 30) / 2 = 25.0
print(ventas.groupby('Tienda').mean())
```

# ***10. Combinación de DataFrames (Merge/Join)***

---

Teoría Mejorada: El método .merge() permite unir tablas basándose en una lógica similar a la de SQL JOINs.

Tipos de unión: inner (intersección), outer (unión total), left (preserva la tabla izquierda) y right (preserva la derecha).

**Codigo de Ejemplo:**

---


```python
df1 = pd.DataFrame({'ID': [1, 2], 'Nombre': ['Ana', 'Luis']})
df2 = pd.DataFrame({'ID': [1, 2], 'Sueldo': [3000, 4000]})

# pd.merge(df1, df2, on='ID'): une por la columna 'ID' (INNER JOIN por defecto)
# Solo incluye filas cuyo ID aparece en AMBAS tablas
# how='left'  → preserva todas las filas de df1 aunque no tengan par en df2
# how='outer' → preserva todos los IDs de ambas tablas (llena con NaN si falta)
print(pd.merge(df1, df2, on='ID'))
```



---



Práctica:** Crea un tercer DataFrame con `ID` y `Ciudad`. Haz un merge con el resultado anterior.


---



```python
df3 = pd.DataFrame({'ID': [1, 2], 'Ciudad': ['Mendoza', 'Rosario']})

# Los merges se encadenan: el resultado de un merge es un DataFrame normal
# que se puede volver a unir con otro df en el mismo campo clave ('ID')
unido = pd.merge(df1, df2, on='ID')          # df1 + df2 → ID, Nombre, Sueldo
print(pd.merge(unido, df3, on='ID'))          # + df3   → ID, Nombre, Sueldo, Ciudad
```


# ***11. Indexación Jerárquica (Multi-Index)***



---



El MultiIndex permite trabajar con datos de dimensiones superiores en un formato bidimensional.

Utilidad: Es ideal para series de tiempo complejas o datos que tienen múltiples niveles de categorización (ej. País -> Ciudad -> Barrio).

**Codigo de Ejemplo:**

---

```python
# MultiIndex: índice con más de un nivel jerárquico
# from_tuples(): construye el índice a partir de una lista de tuplas (nivel1, nivel2)
# Aquí: nivel1 = zona ('CABA'), nivel2 = turno ('M'=mañana, 'T'=tarde)
mi = pd.MultiIndex.from_tuples([('CABA', 'M'), ('CABA', 'T')])
df_m = pd.DataFrame({'Ventas': [100, 150]}, index=mi)

# El DataFrame se muestra con los niveles del índice apilados visualmente
print(df_m)
```


# ***12. Series de Tiempo en Pandas***



---



Pandas nació originalmente para el análisis financiero, por lo que su manejo de fechas es excepcional.

DatetimeIndex: Convierte una columna de tiempo en el índice para permitir selecciones por mes, año o incluso rangos horarios de forma nativa.

Frecuencias: Permite remuestreos (resample) para cambiar de diario a mensual con facilidad.

**Codigo de Ejemplo:**

---





```python
# pd.date_range(): genera una secuencia de fechas con frecuencia fija
# start='20240101' → fecha de inicio (acepta strings 'YYYY-MM-DD' o 'YYYYMMDD')
# periods=5 → cantidad de fechas a generar
# freq='D' → frecuencia diaria
# Otras frecuencias: 'W'=semanal, 'ME'=fin de mes, 'YE'=fin de año, 'H'=hora
fechas = pd.date_range('20240101', periods=5, freq='D')
print(fechas)
```



---


# ***14. Pivot Tables (Tablas Dinámicas)***



---

Una tabla dinámica resume y agrega datos de un DataFrame, transformando filas en columnas para observar relaciones cruzadas.

Flexibilidad: Permite usar múltiples funciones de agregación (aggfunc) simultáneamente, como ['sum', 'mean'].

**Codigo de Ejemplo:**

---

```python
# pivot_table: tabla dinámica que cruza filas y aplica una función de agregación
# values='Monto'   → columna que se va a agregar
# index='Tienda'   → columna que define las filas del resultado
# aggfunc='max'    → función de agregación (también acepta: 'sum', 'mean', 'count', ['sum', 'mean'])
# Tienda A: max(10, 15) = 15  |  Tienda B: max(20, 30) = 30
print(ventas.pivot_table(values='Monto', index='Tienda', aggfunc='max'))
```



---


Práctica:** Genera una tabla dinámica que **cuente** cuántas ventas hubo por tienda.


---



```python
# PRÁCTICA: aggfunc='count' cuenta cuántas filas hay por grupo
# Tienda A aparece 2 veces → count=2  |  Tienda B aparece 2 veces → count=2
print(ventas.pivot_table(values='Monto', index='Tienda', aggfunc='count'))
```
