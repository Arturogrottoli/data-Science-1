# Introducción: NumPy y Pandas

En esta clase vamos a trabajar con dos librerías de Python que son la base de prácticamente todo el ecosistema de Data Science: **NumPy** y **Pandas**. Ambas extienden lo que Python puede hacer de manera nativa, resolviendo problemas que con Python puro serían lentos, verbosos o directamente inmanejables.

---

## ¿Qué es NumPy y para qué sirve?

NumPy (Numerical Python) es una librería para trabajar con datos **numéricos** de forma eficiente. Introduce el `ndarray`, un arreglo de números almacenado directamente en memoria de forma contigua, lo que permite hacer operaciones matemáticas sobre miles o millones de valores a la vez, sin escribir bucles.

**¿Por qué no usar listas de Python directamente?**
Con una lista de Python, para sumarle 10 a cada elemento tenés que recorrerla elemento por elemento. NumPy lo hace internamente en C, mucho más rápido, en una sola instrucción. A medida que los datos crecen, la diferencia es enorme.

**Ejemplos de uso de NumPy:**
- Calcular el promedio de temperatura de cada hora durante un año entero sin escribir un bucle
- Multiplicar los precios de mil productos por un porcentaje de aumento en una sola operación
- Generar una grilla de coordenadas para graficar una función matemática
- Hacer operaciones de álgebra lineal para entrenar un modelo de Machine Learning

---

## ¿Qué es Pandas y para qué sirve?

Pandas es una librería para manipular y analizar **datos tabulares** — el tipo de dato que todos conocemos de Excel o de una base de datos: filas y columnas con distintos tipos de información. Está construida sobre NumPy y agrega etiquetas, manejo de fechas, valores faltantes, agrupamientos y mucho más.

**¿Por qué no usar diccionarios o listas de Python directamente?**
Con Python puro, leer un archivo CSV, filtrar filas, agrupar por categoría y calcular promedios implica decenas de líneas de código con bucles, manejo de errores y conversión de tipos. Pandas hace todo eso en pocas líneas, con operaciones optimizadas y una sintaxis que se lee casi como lenguaje natural.

**Ejemplos de uso de Pandas:**
- Abrir el historial de ventas de una tienda en formato CSV y calcular el total vendido por categoría de producto
- Cruzar la tabla de empleados con la tabla de sueldos usando el número de legajo como nexo
- Detectar qué jugadores de un equipo tienen estadísticas incompletas (goles o minutos faltantes) y decidir cómo tratarlos
- Agrupar las temperaturas diarias de todo un año por mes para ver la evolución estacional

---

## ¿Cuándo uso NumPy y cuándo Pandas?

| Situación | Usá |
|---|---|
| Cálculo numérico puro: promedios, matrices, operaciones matemáticas | NumPy |
| Datos con columnas de distinto tipo, fechas, etiquetas, NaN | Pandas |
| Álgebra lineal, redes neuronales, simulaciones | NumPy |
| Análisis exploratorio, limpieza de datos, reportes | Pandas |

En la práctica, los dos conviven: Pandas usa NumPy internamente, y muchas operaciones de Machine Learning reciben DataFrames de Pandas y los convierten a arrays de NumPy automáticamente.

# ***1. Introducción a Series***

---







Una `pd.Series` es un arreglo **unidimensional etiquetado** capaz de almacenar cualquier tipo de dato. Es la estructura base de Pandas y la columna elemental de un DataFrame.

**Componentes internos:**
- **Datos:** array NumPy subyacente (accesible con `.values`)
- **Índice:** etiquetas para cada elemento (accesible con `.index`)
- **dtype:** tipo de dato único para todos los valores
- **name:** nombre opcional (se convierte en encabezado de columna en un DataFrame)

---

**¿Por qué Series en vez de una lista de Python?**

| Característica | Lista Python | `pd.Series` |
|---|---|---|
| Acceso por nombre | ✗ solo por posición | ✓ por etiqueta (`.loc`) o posición (`.iloc`) |
| Operaciones vectorizadas | ✗ requiere `for` | ✓ automáticas (`+`, `*`, `np.sqrt(s)`) |
| Alineación automática | ✗ | ✓ Pandas alinea por índice al operar entre Series |
| Estadísticas | ✗ | ✓ `.mean()`, `.std()`, `.describe()`, `.value_counts()` |
| Manejo de NaN | ✗ None rompe operaciones | ✓ NaN se propaga de forma controlada |

---

**Sintaxis de creación:**
```
pd.Series(data, index=None, dtype=None, name=None)
```

| Argumento | Qué hace |
|---|---|
| `data` | Lista, dict, ndarray o escalar |
| `index` | Etiquetas personalizadas. Si se omite → `0, 1, 2, ...` |
| `dtype` | Fuerza un tipo de dato (`float32`, `int8`, etc.) |
| `name` | Nombre de la Serie (encabezado de columna en DataFrames) |

---

**Métodos y atributos más usados:**

| Método / Atributo | Descripción |
|---|---|
| `.index` / `.values` | Etiquetas / Datos como ndarray |
| `.dtype` / `.shape` | Tipo de dato / Dimensiones |
| `.loc[etiq]` / `.iloc[pos]` | Acceso por etiqueta / posición |
| `.isna()` / `.notna()` | Máscara booleana de NaN |
| `.value_counts()` | Frecuencia de cada valor único |
| `.apply(func)` | Aplica una función a cada elemento |
| `.sort_values()` | Ordena por valor |
| `.describe()` | Resumen estadístico completo |

**Codigo de Ejemplo:**

---

````python
import pandas as pd

# pd.Series: crea una estructura 1D con índice a partir de cualquier iterable
# index=[...]: asigna etiquetas personalizadas en vez del índice numérico 0,1,2
precios = pd.Series([10.5, 20.0, 15.75], index=['Pan', 'Leche', 'Arroz'])
print(precios)
````

---







El `pd.DataFrame` es la estructura principal de Pandas: una **tabla bidimensional** con ejes etiquetados donde cada columna es una Serie que comparte el mismo índice.

- **Filas (eje 0):** cada fila es una observación / registro
- **Columnas (eje 1):** cada columna es una variable (una `pd.Series`)
- **Índice de filas:** automático `0,1,2...` o personalizado con strings/fechas
- **dtypes por columna:** cada columna puede ser `int64`, `float64`, `object`, `datetime64`, `bool`

---

**Formas de crear un DataFrame:**

| Método | Ejemplo |
|---|---|

| Desde lista de dicts | `pd.DataFrame([{'A':1,'B':3}, {'A':2,'B':4}])` |


# ***3. Creación de Arrays (NumPy)***

---



Los **ndarray** de NumPy son bloques de memoria contigua y homogénea. Son la estructura de datos más eficiente para computación numérica en Python.

**¿Por qué son más rápidos que las listas?**

Las listas de Python almacenan *punteros* a objetos dispersos en memoria. NumPy almacena los valores *directamente* uno al lado del otro en RAM, del mismo tipo. Esto permite enviar operaciones completas a rutinas en **C compilado** sin pasar por el intérprete Python en cada elemento.

| Aspecto | Lista Python | `np.ndarray` |
|---|---|---|
| Almacenamiento | Punteros a objetos en heap | Bloque contiguo en RAM |
| Tipo de datos | Heterogéneo (mix de tipos) | Homogéneo (un único `dtype`) |
| Velocidad operaciones | Lenta (bucle Python) | Rápida (C / BLAS) |
| Uso de memoria | Alto (objetos + punteros) | Bajo (solo datos) |
| Operaciones vectorizadas | ✗ | ✓ |

---

**Broadcasting:** NumPy puede operar entre arrays de **formas distintas** siguiendo reglas automáticas. Por ejemplo, sumar un escalar a una matriz aplica la suma a cada elemento sin copiar el escalar.

```
arr_2d = np.array([[1,2],[3,4]])
arr_2d + 10   →  [[11,12],[13,14]]   # 10 se "expande" sobre toda la matriz
```

---

**Funciones de creación más usadas:**

| Función | Descripción |
|---|---|
| `np.array([1,2,3])` | Desde lista Python |
| `np.arange(start, stop, step)` | Secuencia de números (como `range` pero ndarray) |
| `np.linspace(start, stop, n)` | n puntos equidistantes entre start y stop |
| `np.zeros(shape)` | Matriz de ceros |
| `np.ones(shape)` | Matriz de unos |
| `np.eye(n)` | Matriz identidad n×n |
| `np.random.randint(a,b,size)` | Enteros aleatorios entre a y b |

````python
import numpy as np

# np.array(): convierte una lista Python a ndarray — misma interfaz pero almacenado en C
arr = np.array([1, 2, 3, 4, 5])

# np.arange(start, stop): genera enteros desde start hasta stop-1, igual que range()
# pero devuelve un ndarray listo para operaciones vectorizadas
arr = np.arange(1, 6)

# Operación vectorizada: * 2 multiplica TODOS los elementos sin necesidad de un bucle
# Internamente se ejecuta en C → mucho más rápido que [x*2 for x in lista]
print(arr * 2)
````


# ***4. Generación de Datos y Atributos (NumPy)***

---

NumPy provee métodos para inicializar arrays sin definirlos manualmente. Conocer los **atributos** de un array es el primer paso para entender qué tenemos en memoria.

| `object` | variable | Strings u objetos Python (evitar: pierde velocidad) |

# ***5. Álgebra Lineal con NumPy***

---

NumPy implementa álgebra lineal llamando internamente a **BLAS** (Basic Linear Algebra Subprograms) y **LAPACK** (Linear Algebra PACKage), bibliotecas en Fortran/C extremadamente optimizadas usadas en software científico desde los años 70.

**¿Por qué es importante para Data Science?**

El álgebra lineal es el lenguaje matemático del Machine Learning:
- **Regresión lineal:** `β = (XᵀX)⁻¹ Xᵀy` requiere producto matricial e inversa
- **Redes neuronales:** cada capa es `salida = activacion(W @ entrada + b)`
- **PCA / SVD:** descomposición espectral con `linalg.eig` / `linalg.svd`
- **Sistemas de ecuaciones:** resolver `Ax = b` con `linalg.solve`

---

**Operaciones principales de `np.linalg`:**

| Función | Descripción | Nota |
|---|---|---|
| `A @ B` o `np.dot(A, B)` | Producto matricial | `@` disponible desde Python 3.5 |
| `A.T` | Transpuesta (filas ↔ columnas) | Atributo, no función |
| `np.linalg.inv(A)` | Inversa de A (A × A⁻¹ = I) | Solo para matrices cuadradas no singulares |
| `np.linalg.det(A)` | Determinante | det ≠ 0 → A tiene inversa |
| `np.linalg.solve(A, b)` | Solución de Ax = b | Más eficiente y estable que calcular la inversa |
| `np.linalg.eig(A)` | Valores y vectores propios | Base de PCA |
| `np.linalg.svd(A)` | Descomposición SVD | Compresión de imágenes, reducción dimensional |
| `np.linalg.norm(v)` | Norma (longitud) de un vector | Usado en distancias y regularización |

---

**Producto punto vs producto matricial:**

| Caso | Operación | Resultado |
|---|---|---|
| Vectores 1D | `np.dot(v1, v2)` | Escalar: suma de productos |
| Matrices 2D | `A @ B` | Matriz: combinación lineal de columnas |
| Vector × Matriz | `A @ v` | Vector: transformación lineal de `v` |

El **producto punto** entre dos vectores mide cuánto "apuntan en la misma dirección" (similitud coseno = producto punto / producto de normas).

**Codigo de Ejemplo:**

---


````python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# np.dot(a, b): producto matricial → resultado[i,j] = suma(fila_i(a) × col_j(b))
# Ejemplo: resultado[0,0] = 1*5 + 2*7 = 19
#          resultado[0,1] = 1*6 + 2*8 = 22
# Equivalente moderno y más legible: a @ b (operador @ desde Python 3.5)
print(np.dot(a, b))
````


# ***6. Slices e Indexación en DataFrames***

---

Pandas ofrece dos **indexadores explícitos** para evitar ambigüedades al acceder a datos. Nunca debería usarse `df[número]` directamente para filas porque el comportamiento cambia según el tipo de índice.

**Comparación completa:**

| | `.loc` | `.iloc` |
|---|---|---|
| Accede por | **Etiqueta** del índice | **Posición** entera (0-based) |
| Extremo del slice `[a:b]` | **Incluido** (b entra) | **Excluido** (b no entra, como Python) |
| Acepta | Etiqueta, lista, slice, booleano | Entero, lista, slice entero, booleano |
| Error si no existe | `KeyError` | `IndexError` |
| Tip mnemotécnico | **L**oc → **L**abel | **i**loc → **i**nteger |

---

**Diferencia crítica en slices:**

```
# Suponer índice ['a','b','c','d','e']
df.loc['b':'d']   → filas b, c, d  (d INCLUIDO)
df.iloc[1:4]      → filas en pos 1,2,3  (4 EXCLUIDO)
```

---

**Formas de uso combinado:**

| Sintaxis | Resultado |
|---|---|
| `df.loc['Ana']` | Serie: toda la fila con etiqueta 'Ana' |
| `df.loc['Ana', 'edad']` | Escalar: valor en fila 'Ana', columna 'edad' |
| `df.loc[['Ana','Luis']]` | DataFrame: múltiples filas |
| `df.loc['Ana':'Pedro']` | DataFrame: rango de filas (inclusive) |
| `df.loc[df['edad'] > 30]` | DataFrame: filtro booleano |
| `df.iloc[0]` | Serie: primera fila |
| `df.iloc[0, 2]` | Escalar: fila 0, columna 2 |
| `df.iloc[:5, :3]` | DataFrame: primeras 5 filas y 3 columnas |

---

**Cuándo usar cada uno:**
- Usar **`.loc`** cuando el índice tiene etiquetas significativas (nombres, fechas)
- Usar **`.iloc`** cuando necesitás posición numérica pura (ej: tomar las primeras N filas)

**Codigo de Ejemplo:**

---


````python
# iloc: acceso por POSICIÓN numérica (integer location), siempre 0-indexed
# iloc[0, 1] → fila en posición 0, columna en posición 1 → df tiene Producto en col 0, Stock en col 1
# El resultado es el valor escalar en esa intersección
print(df.iloc[0, 1])
````


# ***7. Operaciones y Funciones sobre DataFrames***

---

Pandas facilita el **Data Wrangling** (transformación y limpieza de datos) mediante operaciones columna por columna que se ejecutan vectorialmente, sin necesidad de bucles.

**Crear y transformar columnas:**

| Operación | Código | Descripción |
|---|---|---|
| Nueva columna calculada | `df['C'] = df['A'] * df['B']` | Multiplicación elemento a elemento |
| Columna condicional | `df['cat'] = df['precio'].apply(lambda x: 'caro' if x > 100 else 'barato')` | Lógica por elemento |
| Renombrar columna | `df.rename(columns={'viejo': 'nuevo'})` | Cambia nombre sin perder datos |
| Eliminar columna | `df.drop(columns=['col'])` | Devuelve copia sin esa columna |
| Cambiar tipo | `df['edad'] = df['edad'].astype(int)` | Conversión de dtype |

---

**Alineación automática de datos:**
Al operar dos DataFrames, Pandas alinea automáticamente por el índice. Las filas que no coincidan quedan como `NaN`. Esto previene errores silenciosos al combinar fuentes distintas.

---

**`.describe()` — ¿qué calcula?**

| Estadística | Descripción |
|---|---|
| `count` | Cantidad de valores NO nulos |
| `mean` | Promedio aritmético |
| `std` | Desviación estándar (dispersión alrededor de la media) |
| `min` / `max` | Valor mínimo y máximo |
| `25%` (Q1) | Primer cuartil: 25% de los datos están por debajo |
| `50%` (mediana) | Valor central de la distribución |
| `75%` (Q3) | Tercer cuartil: 75% de los datos están por debajo |

La diferencia `Q3 - Q1` se llama **rango intercuartil (IQR)** y es una medida robusta de dispersión usada para detectar outliers.

---

**Métodos de transformación más usados:**

| Método | Descripción |
|---|---|
| `.apply(func)` | Aplica una función a cada elemento o fila/columna |


# ***8. Manejo de Valores Faltantes (NaN)***

---

Los valores faltantes aparecen todo el tiempo en datos reales. Algunos ejemplos cotidianos:

- **Equipo de fútbol:** un jugador se lesionó en la fecha 3 y no jugó más. Sus columnas de goles, asistencias y minutos jugados de esas fechas quedan vacías — no es que hizo cero goles, simplemente no hay dato.
- **Catálogo de una tienda:** se cargaron los productos pero no se completó el precio de algunos artículos todavía. El campo `Precio` existe pero está vacío.
- **Verdulería:** el proveedor no siempre informa el peso exacto del cajón. La columna `Peso_kg` tiene datos algunos días y otros no.

En todos estos casos Pandas representa ese "vacío" como `NaN`. El desafío no es solo detectarlo, sino entender **por qué falta** para decidir qué hacer con él.

---

En Pandas, los valores faltantes se representan como **`NaN`** (Not a Number), técnicamente un `float64` especial definido por el estándar IEEE 754. También puede aparecer como `None` (Python) o `NaT` (Not a Time, para fechas).

**¿Por qué aparecen valores faltantes?**
- Datos no registrados en encuestas o formularios
- Errores en la recolección o transmisión de datos
- Joins entre tablas que no tienen coincidencias completas
- Cálculos imposibles (ej: `pct_change()` en la primera fila)

---

**Detección de NaN:**

| Método | Resultado | Uso |
|---|---|---|
| `df.isnull()` | DataFrame booleano (True = NaN) | Ver máscara completa |
| `df.isnull().sum()` | Serie con conteo por columna | Diagnóstico rápido |
| `df.isnull().sum() / len(df)` | Porcentaje de NaN por columna | Decidir estrategia |
| `df.info()` | Muestra non-null count por columna | Vista general rápida |

---

**Estrategias de tratamiento:**

| Estrategia | Cuándo usarla | Riesgo |
|---|---|---|
| `dropna()` — Eliminar | NaN son pocas filas (<5%) y al azar | Perder datos valiosos si hay muchos |
| `fillna(media)` — Imputar con estadístico | Variable numérica, distribución simétrica | Reduce varianza artificialmente |
| `fillna(mediana)` — Imputar con mediana | Variable numérica con outliers | Más robusta que la media |
| `fillna(moda)` — Imputar con moda | Variable categórica | Solo si la categoría más frecuente es representativa |
| `fillna(method='ffill')` — Propagación | Series de tiempo (precio, temperatura) | Asume que el valor anterior sigue vigente |
| Imputación KNN / ML | Muchos NaN, variable importante | Más precisa pero más compleja |

---

**Parámetros útiles de `dropna()`:**

| Parámetro | Opciones | Descripción |
|---|---|---|
| `how` | `'any'` (default) / `'all'` | Eliminar si hay AL MENOS un NaN / solo si TODOS son NaN |
| `subset` | lista de columnas | Solo revisar NaN en esas columnas |
| `axis` | `0` (filas) / `1` (columnas) | Eliminar filas o columnas |

**Codigo de Ejemplo:**

---



````python
import numpy as np

# np.nan: valor especial float que Pandas usa para representar datos ausentes
# Al crearlo con np.nan, Pandas lo reconoce automáticamente como "faltante"
df_nulo = pd.DataFrame({'Ventas': [100, np.nan, 200]})

# fillna(valor): devuelve una COPIA con todos los NaN reemplazados por el valor dado
# Aquí reemplazamos NaN por 0; también se puede usar la media: fillna(df_nulo['Ventas'].mean())
print(df_nulo.fillna(0))
````


# ***9. Agrupamiento de Datos (Group By)***

---

Imaginate que tenés la planilla de ventas de una cadena de supermercados con miles de filas mezcladas: cada fila es una venta, con la sucursal, el producto y el monto. Si querés saber cuánto vendió cada sucursal en total, no podés sumar todo junto — primero tenés que *agrupar* por sucursal y recién ahí sumar. Eso es exactamente lo que hace GroupBy: junta las filas que tienen algo en común y les aplica un cálculo. Es el mismo concepto que cuando un DT separa a los jugadores por posición para calcular el promedio de goles de los delanteros por separado del de los defensores.

---

El **GroupBy** de Pandas implementa el patrón **Split → Apply → Combine**, que es la base de prácticamente todo análisis agregado de datos.

**El flujo Split → Apply → Combine:**

```
DataFrame original
       ↓
  SPLIT: dividir en grupos según una columna (ej: 'Tienda')
       ↓
  APPLY: aplicar una función a cada grupo (ej: sum, mean, count)
       ↓
  COMBINE: reunir los resultados en un nuevo DataFrame
```

---

**Funciones de agregación disponibles:**

| Función | Descripción |
|---|---|
| `.sum()` | Suma de los valores del grupo |
| `.mean()` | Promedio |
| `.median()` | Mediana |
| `.count()` | Cantidad de valores no nulos |
| `.min()` / `.max()` | Mínimo y máximo |
| `.std()` / `.var()` | Desviación estándar / varianza |
| `.first()` / `.last()` | Primer / último valor del grupo |
| `.agg(['mean', 'std'])` | Varias funciones a la vez |
| `.agg({'col1': 'sum', 'col2': 'mean'})` | Función diferente por columna |

---

**Variantes de groupby:**

| Código | Descripción |
|---|---|
| `df.groupby('col')['val'].mean()` | Agrupar por una columna, agregar una |
| `df.groupby(['col1','col2']).sum()` | Agrupar por múltiples columnas (MultiIndex) |
| `df.groupby('col').transform('mean')` | Reemplaza cada valor por la media del grupo (mismo shape que original) |
| `df.groupby('col').filter(lambda g: len(g) > 2)` | Mantener solo grupos que cumplen una condición |

---

**`transform` vs `agg`:**

| | `.agg()` | `.transform()` |
|---|---|---|
| Shape resultado | Reducido (un valor por grupo) | Igual al original (una fila por registro) |
| Uso típico | Resumen estadístico | Normalización por grupo, completar NaN con media del grupo |

**Codigo de Ejemplo:**

---






````python
# DataFrame de ventas con dos tiendas y cuatro registros
ventas = pd.DataFrame({'Tienda': ['A', 'B', 'A', 'B'], 'Monto': [10, 20, 15, 30]})

# groupby('Tienda'): SPLIT — divide el df en grupos por valor único de 'Tienda'
# .sum(): APPLY — suma los montos dentro de cada grupo
# COMBINE — devuelve un nuevo df con un renglón por grupo
# Tienda A: 10 + 15 = 25  |  Tienda B: 20 + 30 = 50
print(ventas.groupby('Tienda').sum())
````

# ***10. Combinación de DataFrames (Merge/Join)***

---

El método `.merge()` permite combinar dos DataFrames basándose en una o más columnas clave, replicando la lógica de los **JOINs de SQL**.

**Tipos de unión (`how=`):**

| Tipo | Descripción | Registros incluidos |
|---|---|---|
| `'inner'` (default) | Intersección | Solo filas con clave en AMBAS tablas |
| `'left'` | Left join | Todas las filas del df izquierdo; NaN si no hay par derecho |
| `'right'` | Right join | Todas las filas del df derecho; NaN si no hay par izquierdo |
| `'outer'` | Full outer join | Todas las filas de ambas tablas; NaN donde no haya par |

---

**Analogía visual (con ID como clave):**

```
df1: ID=[1,2,3]    df2: ID=[2,3,4]

inner  → ID=[2,3]         (solo los que están en ambos)
left   → ID=[1,2,3]       (todos de df1; ID=1 tendrá NaN en cols de df2)
right  → ID=[2,3,4]       (todos de df2; ID=4 tendrá NaN en cols de df1)
outer  → ID=[1,2,3,4]     (todos; NaN donde no haya par)
```

---

**Parámetros de `pd.merge()`:**

| Parámetro | Descripción |
|---|---|
| `on='col'` | Columna clave presente en AMBOS DataFrames |
| `left_on` / `right_on` | Claves con distinto nombre en cada tabla |
| `how` | Tipo de join: `'inner'`, `'left'`, `'right'`, `'outer'` |
| `suffixes=('_x','_y')` | Sufijos para columnas duplicadas (no la clave) |
| `validate` | Verifica tipo de relación: `'1:1'`, `'1:m'`, `'m:1'` |

---

**Diferencia entre `merge` y `concat`:**

| | `pd.merge()` | `pd.concat()` |
|---|---|---|
| Une por | Columna(s) clave compartida | Posición (apila filas o columnas) |
| Analogía | SQL JOIN | UNION ALL |
| Uso | Tablas relacionadas (tablas de BD) | Agregar más filas del mismo dataset |

**Codigo de Ejemplo:**

---


````python
df1 = pd.DataFrame({'ID': [1, 2], 'Nombre': ['Ana', 'Luis']})
df2 = pd.DataFrame({'ID': [1, 2], 'Sueldo': [3000, 4000]})

# pd.merge(df1, df2, on='ID'): une por la columna 'ID' (INNER JOIN por defecto)
# Solo incluye filas cuyo ID aparece en AMBAS tablas
# how='left'  → preserva todas las filas de df1 aunque no tengan par en df2
# how='outer' → preserva todos los IDs de ambas tablas (llena con NaN si falta)
print(pd.merge(df1, df2, on='ID'))
````


# ***11. Indexación Jerárquica (Multi-Index)***



---



El **MultiIndex** (índice jerárquico) permite representar datos de más de dos dimensiones en una tabla 2D, evitando duplicar columnas o crear múltiples DataFrames separados.

**¿Cuándo usarlo?**
- Datos con múltiples niveles de categorización: País → Ciudad → Barrio
- Series de tiempo con múltiples entidades: Empresa × Fecha
- Resultados de `groupby` con múltiples columnas agrupadas

---

**Formas de crear un MultiIndex:**

| Método | Código |
|---|---|
| Desde lista de tuplas | `pd.MultiIndex.from_tuples([('A','x'), ('A','y'), ('B','x')])` |
| Desde arrays separados | `pd.MultiIndex.from_arrays([['A','A','B'], ['x','y','x']])` |
| Producto cartesiano | `pd.MultiIndex.from_product([['A','B'], ['x','y']])` |
| Resultado de groupby | `df.groupby(['col1','col2']).sum()` genera MultiIndex automáticamente |

---

**Acceso a datos con MultiIndex:**

| Código | Qué devuelve |
|---|---|
| `df.loc['CABA']` | Todas las filas del nivel 1 = 'CABA' (elimina ese nivel) |
| `df.loc[('CABA', 'T')]` | Fila exacta: nivel1='CABA', nivel2='T' |
| `df.loc['CABA', 'T']` | Equivalente al anterior |
| `df.xs('T', level=1)` | Cross-section: todas las filas donde nivel2='T' |
| `df.reset_index()` | Convierte el MultiIndex en columnas normales |

---

**Niveles del MultiIndex:**

| Atributo | Descripción |
|---|---|
| `df.index.names` | Nombres de cada nivel |
| `df.index.levels` | Valores únicos de cada nivel |
| `df.index.get_level_values(0)` | Todos los valores del nivel 0 |
| `df.swaplevel()` | Intercambia el orden de los niveles |
| `df.sort_index()` | Ordena el MultiIndex (necesario para slicing eficiente) |

**Codigo de Ejemplo:**


# ***12. Series de Tiempo en Pandas***



---



Pandas nació originalmente para el **análisis financiero** (fue desarrollado por Wes McKinney en AQR Capital Management). Por eso su manejo de fechas y series de tiempo es excepcionalmente potente.

**¿Qué es un DatetimeIndex?**
Es un índice basado en timestamps (`datetime64`). Al establecerlo, Pandas desbloquea operaciones temporales nativas: seleccionar por año/mes, hacer resampling, calcular diferencias entre fechas.

---

**Creación de rangos de fechas (`pd.date_range`):**

| Parámetro | Descripción | Ejemplo |
|---|---|---|
| `start` | Fecha de inicio | `'2024-01-01'` |
| `end` | Fecha de fin | `'2024-12-31'` |
| `periods` | Cantidad de fechas | `12` |
| `freq` | Frecuencia | Ver tabla abajo |

**Frecuencias más comunes:**

| Código | Frecuencia |
|---|---|
| `'D'` | Diaria |
| `'W'` | Semanal (domingos) |
| `'ME'` | Fin de mes |
| `'QE'` | Fin de trimestre |
| `'YE'` | Fin de año |
| `'H'` | Hora |
| `'T'` o `'min'` | Minuto |
| `'B'` | Días hábiles (lunes a viernes) |

---

**Operaciones temporales más útiles:**

| Operación | Código | Descripción |
|---|---|---|
| Seleccionar por año | `df.loc['2024']` | Todas las fechas de 2024 |
| Seleccionar por mes | `df.loc['2024-03']` | Todo marzo 2024 |
| Rango de fechas | `df.loc['2024-01':'2024-06']` | Enero a junio 2024 |
| Resampling | `df.resample('ME').mean()` | Agrupa por mes y calcula media |
| Shift temporal | `df.shift(1)` | Desplaza los valores 1 período (genera NaN en la primera fila) |
| Diferencia | `df.diff(1)` | Diferencia absoluta entre períodos consecutivos |
| Retorno % | `df.pct_change()` | Variación porcentual entre períodos |

Un número en una tabla no dice casi nada hasta que lo ves graficado. Si tenés las ventas de los últimos doce meses escritas en filas, es difícil ver si subieron, bajaron o tuvieron un pico en diciembre. Un gráfico de línea lo muestra en un segundo. Pandas te permite graficar directamente desde cualquier DataFrame o Serie sin necesidad de preparar los datos manualmente — en una sola línea ya tenés el gráfico listo en el notebook.

---

**Codigo de Ejemplo:**

---





````python
# pd.date_range(): genera una secuencia de fechas con frecuencia fija
# start='20240101' → fecha de inicio (acepta strings 'YYYY-MM-DD' o 'YYYYMMDD')
# periods=5 → cantidad de fechas a generar
# freq='D' → frecuencia diaria
# Otras frecuencias: 'W'=semanal, 'ME'=fin de mes, 'YE'=fin de año, 'H'=hora
fechas = pd.date_range('20240101', periods=5, freq='D')
print(fechas)
````


# ***14. Pivot Tables (Tablas Dinámicas)***



---

Una **Pivot Table** (tabla dinámica) resume y agrega datos de un DataFrame reorganizando la información: una columna se convierte en filas, otra en columnas, y los valores se agregan con una función. Es idéntica a la tabla dinámica de Excel.

**¿Cómo funciona?**

```
DataFrame largo (tidy):           Pivot Table:
Tienda | Mes  | Monto             Mes →   Ene   Feb
  A      Ene    10         →    Tienda A   10    30
  A      Feb    30               Tienda B   20    15
  B      Ene    20
  B      Feb    15
```

---

**Parámetros de `pivot_table()`:**

| Parámetro | Descripción |
|---|---|
| `values` | Columna a agregar (los números de la tabla) |
| `index` | Columna que define las FILAS de la tabla |
| `columns` | Columna que define las COLUMNAS de la tabla (opcional) |
| `aggfunc` | Función de agregación: `'sum'`, `'mean'`, `'count'`, `'max'`, `['sum','mean']` |
| `fill_value` | Valor para reemplazar NaN en el resultado |
| `margins=True` | Agrega fila/columna "All" con el total general |

---

**Diferencias con `groupby`:**

| | `groupby().agg()` | `pivot_table()` |
|---|---|---|
| Resultado | DataFrame largo (tidy) | DataFrame ancho (tabla cruzada) |
| Legibilidad | Mejor para más procesamiento | Mejor para visualización humana |
| Múltiples funciones | `agg(['sum','mean'])` → MultiIndex columnas | `aggfunc=['sum','mean']` → MultiIndex columnas |
| Columnas cruzadas | ✗ | ✓ con parámetro `columns=` |

---

**Uso avanzado:**

| Ejemplo | Descripción |
|---|---|
| `aggfunc=['sum', 'count']` | Dos estadísticas a la vez |
| `margins=True` | Agrega subtotales por fila y columna |
| `fill_value=0` | Reemplaza NaN por 0 en celdas sin datos |

**Codigo de Ejemplo:**

---

````python
# pivot_table: tabla dinámica que cruza filas y aplica una función de agregación
# values='Monto'   → columna que se va a agregar
# index='Tienda'   → columna que define las filas del resultado
# aggfunc='max'    → función de agregación (también acepta: 'sum', 'mean', 'count', ['sum', 'mean'])
# Tienda A: max(10, 15) = 15  |  Tienda B: max(20, 30) = 30
print(ventas.pivot_table(values='Monto', index='Tienda', aggfunc='max'))
````


# ***15. Lectura y Escritura de Archivos***



---

Pandas actúa como un **puente universal** entre casi cualquier fuente de datos y el análisis en Python. Cada formato tiene su función de lectura y escritura simétrica.

**Formatos soportados:**

| Formato | Lectura | Escritura | Librería extra |
|---|---|---|---|
| CSV / TXT | `pd.read_csv(path)` | `df.to_csv(path)` | — |
| Excel (.xlsx) | `pd.read_excel(path)` | `df.to_excel(path)` | `openpyxl` |
| JSON | `pd.read_json(path)` | `df.to_json(path)` | — |
| SQL | `pd.read_sql(query, conn)` | `df.to_sql(table, conn)` | `sqlalchemy` |
| Parquet | `pd.read_parquet(path)` | `df.to_parquet(path)` | `pyarrow` |
| URL / API | `pd.read_csv(url)` | — | — |
| HTML (tablas web) | `pd.read_html(url)` | `df.to_html(path)` | `lxml` |

---

**Parámetros clave de `read_csv`:**

| Parámetro | Descripción | Ejemplo |
|---|---|---|
| `sep` | Separador de columnas | `sep=';'` para CSV europeo |
| `index_col` | Columna que actúa como índice | `index_col=0` o `index_col='fecha'` |
| `header` | Fila con nombres de columna | `header=0` (default) o `header=None` |
| `names` | Nombres de columnas si no hay header | `names=['a','b','c']` |
| `usecols` | Solo cargar ciertas columnas | `usecols=['fecha','precio']` |
| `nrows` | Leer solo las primeras N filas | `nrows=1000` (exploración rápida) |
| `chunksize` | Leer de a partes (archivos enormes) | `chunksize=10000` |
| `encoding` | Codificación del archivo | `encoding='latin-1'` para archivos en español |
| `dtype` | Forzar tipo de columnas | `dtype={'codigo': str}` |
| `parse_dates` | Convertir columnas a datetime | `parse_dates=['fecha']` |

---

**¿Cuándo usar `chunksize`?**
Cuando el archivo no cabe en la RAM. En lugar de un DataFrame, `read_csv(..., chunksize=N)` devuelve un iterador de DataFrames de N filas cada uno, que se procesan de a uno.

```python
for chunk in pd.read_csv('enorme.csv', chunksize=10000):
    procesar(chunk)   # cada chunk es un DataFrame de 10.000 filas
```
