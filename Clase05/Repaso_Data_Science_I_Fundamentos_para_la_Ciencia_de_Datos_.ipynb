# 📅 Repaso Semana 2: Controlar el Flujo y Automatizar

Acá aprendemos a ponerle "cerebro" al script: que tome caminos distintos según la situación (if/else) y que procese listas de datos sin que tengamos que escribir una línea por cada uno (for).

```python
print("\n--- REPASO SEMANA 2 ---")

# Lista de temperaturas recolectadas en el día
historial_temps = [72.0, 85.3, 91.0, 68.4, 88.9]

# Queremos separar las temperaturas que son "Peligrosas" (mayores a 80)
temps_altas = []

for temp in historial_temps:
    # El bucle agarra cada 'temp' de la lista una por una
    if temp > 80.0:
        # Si cumple la condición, la guardamos en nuestra lista vacía
        temps_altas.append(temp)
    else:
        # Si no la cumple, pasa de largo
        pass

print(f"Todas las mediciones: {historial_temps}")
print(f"Mediciones que superaron el peligro (>80): {temps_altas}")
```

# 📅 Repaso Semana 3: NumPy (Cálculos en Bloque)

Hacer cuentas matemáticas con listas nativas de Python es lento y requiere muchos bucles for. NumPy inventa el array (arreglo), que te permite hacerle la misma cuenta a millones de datos a la vez (vectorización).

```python
import numpy as np
print("\n--- REPASO SEMANA 3 ---")

# Imaginemos que tenemos los precios de 4 productos en dólares
precios_usd = np.array([10.0, 25.5, 100.0, 5.25])

# Queremos pasarlos a pesos argentinos (suponiendo un valor ficticio de $1000)
tipo_cambio = 1000

# En Python común tendrías que hacer un 'for'. En NumPy multiplicás directo:
precios_ars = precios_usd * tipo_cambio

print(f"Precios en USD: {precios_usd}")
print(f"Precios transformados a ARS en un solo paso: {precios_ars}")

# También podés filtrar usando condiciones directo en los corchetes (Máscaras)
precios_caros = precios_usd[precios_usd > 20.0]
print(f"Productos que salen más de 20 USD: {precios_caros}")
```

# 📅 Repaso Semana 4: Pandas (El Excel de Python)

Pandas agarra los arreglos de NumPy y les pone filas y columnas, creando un DataFrame. Es la herramienta clave para limpiar tablas, agrupar datos y manejar fechas.

```python
import pandas as pd
import numpy as np
print("\n--- REPASO SEMANA 4 ---")

# Creamos una tabla manual para entender su anatomía
datos_ventas = {
    "Producto": ["Teclado", "Mouse", "Teclado", "Monitor", "Mouse"],
    "Unidades": [10, 15, np.nan, 5, 20], # np.nan representa un dato faltante (vacío)
    "Precio_Unitario": [50, 25, 50, 300, 25]
}

df = pd.DataFrame(datos_ventas)
print("Tabla original (con un dato faltante en fila 2):")
print(df)

# 1. Limpieza: Llenamos el dato faltante (Fila de unidades de teclado) con un 0
df["Unidades"] = df["Unidades"].fillna(0)

# 2. Transformación: Creamos una columna calculada
df["Total_Venta"] = df["Unidades"] * df["Precio_Unitario"]

# 3. Agrupación: ¿Cuánto dinero total recaudó cada tipo de producto?
reporte_final = df.groupby("Producto")["Total_Venta"].sum()

print("\nReporte total recaudado por producto:")
print(reporte_final)
```
