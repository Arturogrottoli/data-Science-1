🟢 Unidad 1 a 3: Del Python Puro a NumPy y Pandas
Enfocado en la transición de listas hacia estructuras de alto rendimiento.

```python
import numpy as np
import pandas as pd

# 1. Patrón Idiomático (List Comprehension)
datos_crudos = [10, 50, 30, 40, 50]
procesados = [x * 1.21 for x in datos_crudos] # Cálculo de IVA rápido

################### Mucho tiempo necesitamos velocidad####################
datos_crudos = [10, 20, 30, 40, 50]
procesados = [] # 1. Primero creamos una lista vacía

for x in datos_crudos: # 2. Empezamos a recorrer la lista
    resultado = x * 1.21 # 3. Hacemos el cálculo
    procesados.append(resultado) # 4. Agregamos el resultado a la lista
#########################################################################3

# 2. NumPy: El poder del Broadcasting
# En lugar de bucles, operamos sobre todo el arreglo a la vez
arr = np.array(datos_crudos)
arr_iva = arr * 1.21

# 3. Pandas: Carga y Selección
# Creamos un DF rápido para practicar
df = pd.DataFrame({
    'Ventas': arr,
    'Con_IVA': arr_iva,
    'Categoria': ['A', 'B', 'A', 'C', 'B']
})

# Selección y Transformación rápida
ventas_altas = df[df['Ventas'] > 25] # Filtrado
resumen = df.groupby('Categoria')['Ventas'].sum() # Agregación
```

🔵 Unidad 5 a 10: Visualización (De Estático a Interactivo)
Comparativa entre el estándar de la industria (Matplotlib) y la interactividad moderna (Plotly).

```python
import matplotlib.pyplot as plt
import plotly.express as px

# 6. Matplotlib (Gráfico estático para reportes)
plt.figure(figsize=(8, 4))
plt.plot(df['Ventas'], marker='o', color='teal', label='Tendencia')
plt.title("Ventas por Período")
plt.legend()
plt.show()

# 10. Plotly (Gráfico interactivo para dashboards)
fig = px.bar(df, x='Categoria', y='Ventas', color='Categoria', title="Ventas por Categoría")
fig.show()
```

🟡 Unidad 7 a 11: Funciones Robustas y Errores
Diseño de código profesional y manejo de excepciones.

```python
# 7. Funciones con argumentos y manejo de errores (11)
def calcular_metricas(datos, multiplicador=1.0):
    try:
        if not isinstance(datos, (list, np.ndarray)):
            raise ValueError("Los datos deben ser una lista o arreglo.")
        return np.mean(datos) * multiplicador
    except Exception as e:
        return f"Error en el cálculo: {e}"

# 8. Testeo rápido
print(f"Test 1 (Correcto): {calcular_metricas([10, 20, 30])}")
print(f"Test 2 (Error): {calcular_metricas('No soy una lista')}")
```
